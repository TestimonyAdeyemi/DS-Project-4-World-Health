```python
# Set-up
import pandas as pd
from pandas.io.parsers.readers import read_csv
#cancer = pd.DataFrame("30-70cancerChdEtc.csv")
cancer = pd.read_csv('DataSets/30-70cancerChdEtc.csv')
cancer.head(5)
```

```python
#STEP 1: I want to describe the data
```

```python
print(cancer.head())
print('Number of Countries in our Data Set:' + str(cancer['Location'].nunique()))
```

```python
# STEP 2: We want to identfy questions we can answer with our data

# Percentage increase/decrease of cancer in each the country from 2010 to 2016 for both sexes
# Percentage increase/decrease of cancer in each the country from 2010 to 2016 for males
# Percentage increase/decrease of cancer in each the country from 2010 to 2016 for females
# The countries with the highest increase in cancer probabilites
# The countries with the highest decrease in cancer probabilites
```

```python
#Question 1

#filter out males and females and save it in a new datafram.
#rename the period column in each df to 2010, 2015, 2016


cancer2016 = cancer[(cancer.Dim1=="Both sexes") & (cancer.Period == 2016)]
cancer2016.columns = cancer2016.columns.str.replace('First Tooltip', '2016')
cancer2016 = cancer2016.drop(['Period', 'Indicator', 'Dim1'], axis=1)
print(cancer2016)

cancer2010 = cancer[(cancer.Dim1=="Both sexes") & (cancer.Period == 2010)]
cancer2010.columns = cancer2010.columns.str.replace('First Tooltip', '2010')
cancer2010 = cancer2010.drop(['Period', 'Indicator', 'Dim1'], axis=1)
print(cancer2010)


cancer2015 =  cancer[(cancer.Dim1=="Both sexes") & (cancer.Period == 2015)]
cancer2015.columns = cancer2015.columns.str.replace('First Tooltip', '2015')
cancer2015 =  cancer2015.drop(['Period', 'Indicator', 'Dim1'], axis=1)
print(cancer2015)
```

```python
#then merge all df to have 2010,2015 and 2016 in one df
cancer_df = pd.merge(cancer2010, cancer2015, on=['Location'])
```

```python
cancer_df = pd.merge(cancer_df, cancer2016, on=['Location'])
```

```python
cancer_df.head(10)
```

```python
#calculate the difference bettween 2010 and 2015
cancer_df['change'] = cancer_df.apply(lambda row: row['2016'] - row['2010'], axis=1)
cancer_df
#cancer_df['change'] = cancer_df[2010].sub(cancer_df[2016])
```

```python
cancer_arranged = cancer_df.sort_values(by=['change'], ascending=False)
print("Top countries with the highest increase in cancer probabilities:")
print(cancer_arranged.head(5))
print("\n Top countries with the highest decrease in cancer probabilities:")
print(cancer_arranged.tail(5))
```

## Connecting another data set to figure out what could be causing the increase and descrease in the propbablity of deaths by non communicable diseases

```python
AlcoholAbuse = pd.read_csv('DataSets/alcoholSubstanceAbuse.csv')
AlcoholAbuse = AlcoholAbuse[(AlcoholAbuse.Dim1=="Both sexes") & (AlcoholAbuse.Period == 2015)]
AlcoholAbuse.head(10)
```

```python
cancer_abuse = pd.merge(cancer_df, AlcoholAbuse, on=['Location'])
```

```python
cancer_abuse.head(10)
```

```python
cancer_abuse.columns = cancer_abuse.columns.str.replace('First Tooltip', 'AlcoholAbuse')
cancer_abuse = cancer_abuse.drop(['Period', 'Indicator', 'Dim1'], axis=1)
print(cancer_abuse)
```

```python
ax1 = cancer_abuse.plot.scatter(x='2016',
                      y='AlcoholAbuse',
                      c='DarkBlue')

import numpy as np

# Correlation coefficients (r values)
r = np.corrcoef(cancer_abuse['2'], cancer_abuse['AlcoholAbuse'])[1,0]
print('Correlation coefficient for cancer probabity and  alcohol abuse is :', r)

r = np.corrcoef(cancer_abuse['2'], cancer_abuse['AlcoholAbuse'])[1,0]
print('Correlation coefficient for cancer probabity and  alcohol abuse is :', r)
```

```python
cancer_abuse.plot.scatter('2015', 'AlcoholAbuse')
a, b = np.polyfit(cancer_abuse['2015'], cancer_abuse['AlcoholAbuse'], 1)  # Regression line is y = ax + b
x1 = min(cancer_abuse['2015'])
x2 = max(cancer_abuse['2015'])
plt.plot([x1, x2], [a * x1 + b, a * x2 + b], color='red')
plt.show()
```
